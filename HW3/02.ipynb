```python
# Alexis Ayuso
# 410921355
# Introduction to Big Data Analytics
# HW3: Q2 - to show the box plots

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

ff = pd.read_csv('forestfires.csv')
ff.head() # displays the first five rows of the dataframe by default
```

```python
# show boxplots of "FFMC" for each day in the week using Pandas library
ff.boxplot(column='FFMC', by='day', figsize=(8,5))
plt.show()
```

```python
# boxplots of "DMC" for each day in the week using Matplotlib library
'''
plt.boxplot(ff['DMC']) # to plot DMC (single column)
plt.show()
'''
fri = ff.loc[ff['day']=='fri', 'DMC'].values
tue = ff.loc[ff['day']=='tue', 'DMC'].values
sat = ff.loc[ff['day']=='sat', 'DMC'].values
sun = ff.loc[ff['day']=='sun', 'DMC'].values
mon = ff.loc[ff['day']=='mon', 'DMC'].values
wed = ff.loc[ff['day']=='wed', 'DMC'].values
thu = ff.loc[ff['day']=='thu', 'DMC'].values
plt.boxplot([fri, tue, sat, sun, mon, wed, thu])
```

```python
# boxplots of "DC" for each day in the week using Seaborn library
sb.boxplot(data=ff, x='day', y='DC')
plt.show()
```
