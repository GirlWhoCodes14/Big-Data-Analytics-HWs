```python
# Alexis Ayuso
# 410921355
# Introduction to Big Data Analytics
# HW3: Q3 - to draw data points on a plot for a large dataset

import pandas as pd, colorcet
import datashader as ds
import datashader.transfer_functions as tf
from datashader.utils import export_image

df = pd.read_parquet("osm-1billion.parq",  engine='pyarrow') # to read the data set
df.size # to get the size of the dataset
```

```python
df.head() # displays the first five rows of the dataframe by default
```

```python
# use the Datashader to understand and plot the dataset
cvs = ds.Canvas(plot_width=850, plot_height=850) # Create a canvas
agg = cvs.points(df, "x", "y") # Aggregate the data
img = tf.shade(agg, cmap=colorcet.fire, how='log') # Transfer function
```

```python
export_image(img, filename='03_image', fmt=".png",  export_path=".")
```
