You need to use Scatter plot for each pairing of the variables and plot Kernel density estimate (kdf) for the marginal plots along the diagonal:

```python
# Alexis Ayuso
# 410921355
# Introduction to Big Data Analytics
# HW3: Q1 - to show the pair plots

import seaborn as sb

sb.set_theme(style="whitegrid") 
iris = sb.load_dataset("iris")
```

```python
iris.head() # displays the first five rows of the dataframe by default
```

```python
sb.pairplot(iris, diag_kind="kde", hue="species", height=3, aspect=1.2) # use the library Seaborn’s methods: Pair Plot
```

```python
g = sb.PairGrid(iris, hue="species", diag_sharey=False, height=3, aspect=1.2) # use the library Seaborn’s method: Pair Grid
g.map_offdiag(sb.scatterplot) # use Scatter plot for each pairing of the variables
g.map_diag(sb.kdeplot, fill=True) # plot Kernel density estimate (kdf) for the marginal plots along the diagonal
g.add_legend()
```

Reference: seaborn.pairplot (https://seaborn.pydata.org/generated/seaborn.pairplot.html)

Plot pairwise relationships in a dataset.

By default, this function will create a grid of Axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column.

This is a high-level interface for PairGrid that is intended to make it easy to draw a few common styles. You should use PairGrid directly if you need more flexibility.

Reference : Seaborn - Kernel Density Estimates (https://www.tutorialspoint.com/seaborn/seaborn_kernel_density_estimates.htm)

Scatter plot is the most convenient way to visualize the distribution where each observation is represented in two-dimensional plot via x and y axis.

Kernel Density Estimation (KDE) is a way to estimate the probability density function of a continuous random variable (i.e. the distribution of a variable). It is used for non-parametric analysis.
